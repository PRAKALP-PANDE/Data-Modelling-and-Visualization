```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

df = pd.read_csv('/Users/sonavaneonkar/Downloads/Data/AQI Data Set.csv', parse_dates=['Mounths'])
df.head()
```

```python
df.columns
```

```python
column_names = ['ID', 'Months', 'PM10', 'SO2', 'NOx', 'PM25', 'NH3', 'O3', 'CO', 'Benzene', 'AQI']

df.columns = column_names
df.head()
```

```python
df.isna().sum()
```

```python
df.dropna(inplace = True)
df.isna().sum()
```

```python
df.describe()
```

```python
df.info()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df['Months'], df['AQI'])

plt.xlabel('Date')
plt.ylabel('AQI')
plt.xticks(rotation=75) 

plt.title('Air Quality Index A(Q)I Ternd Over Time') 
plt.show()
```

```python
plt.figure(figsize = (12, 6))
plt.plot(df['Months'], df['PM10'], color = 'red')
plt.xlabel('Date')
plt.xticks(rotation = 75)
plt.ylabel('PM10')
```

```python
plt.figure(figsize = (12, 6))
plt.plot(df['Months'], df['PM25'], color = 'blue')
plt.xlabel('Date')
plt.xticks(rotation = 75)
plt.ylabel('PM2.5')
```

```python
plt.figure(figsize = (12, 6))
plt.plot(df['Months'], df['CO'], color = 'blue')
plt.xlabel('Date')
plt.xticks(rotation = 75)
plt.ylabel('Carbon Monoxide Levels')
```

```python
plt.figure(figsize = (12, 6))
plt.plot(df['Months'], df['SO2'], color = 'blue')
plt.xlabel('Date')
plt.xticks(rotation = 75)
plt.ylabel('SO2')
```

```python
plt.figure(figsize = (12, 6))
plt.plot(df['Months'], df['NOx'], color = 'blue')
plt.xlabel('Date')
plt.xticks(rotation = 75)
plt.ylabel('NOx')
```

```python
plt.figure(figsize = (12, 6))
plt.bar(df['Months'], df['AQI'], color = 'blue', label = "AQI")
plt.title("AQI comparison Across months")
plt.xlabel('Months')
plt.xticks(rotation = 75)
plt.ylabel('AQI value')
plt.legend()

plt.show()
```

```python
plt.figure(figsize = (12, 6))
df.boxplot(color = 'blue')
plt.title("Distribution of AQI values for Pollutants")
plt.show()
```

```python
plt.figure(figsize = (12, 6))
plt.scatter(df["AQI"], df["PM25"], c = "red", label = "PM2.5")
plt.scatter(df["AQI"], df["SO2"], c = "blue", label = "SO2")
plt.scatter(df["AQI"], df["O3"], c = "green", label = "O3")
plt.title("Scatter Plot: AQi vs. Pollutant Levels")
plt.xlabel("AQI Value")
plt.ylabel("Pollutant Level")
plt.legend()
plt.grid(True)
plt.show()
```
